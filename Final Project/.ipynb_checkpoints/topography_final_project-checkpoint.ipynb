```python
import numpy as np
import matplotlib.pyplot as plt
from array import array
```

```python
dirin ='/home/mwestal/CLIM-470-Numerical-Weather-Prediction/Final Project/'
fil = 'topography_high_res.dat'
Nx = 49
Ny = 17
```

```python
recl = 4*Nx*Ny
ds_bin = open(dirin+fil,'rb')
tmp = ds_bin.read(recl)
hs=np.reshape(array('f',tmp),(Ny, Nx))
```

```python
np.shape(hs)
plt.plot(hs[0,:],'-o')
plt.xlabel('Nx gridpoints', fontsize=16)
plt.ylabel('hs surface height in m', fontsize=16)
plt.suptitle('Shallow Water Model Bottom Topography, d = 125km', fontsize=20)
```
