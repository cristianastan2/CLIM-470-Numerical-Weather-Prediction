```python
import numpy as np
from array import array
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import matplotlib.patches as mpatches
```

```python
dirin ='/home/mwestal/CLIM-470-Numerical-Weather-Prediction/Final Project/'

u0lowfil = 'u_initial_low_res.dat'
v0lowfil = 'v_initial_low_res.dat'
h0lowfil = 'h_initial_low_res.dat'

Nx = 13
Ny = 5
```

```python
recl = 4*Nx*Ny
u0l_bin = open(dirin+u0lowfil, 'rb')
tmpu0l = u0l_bin.read(recl)
u0l = np.reshape(array('f',tmpu0l),(Ny,Nx))
```

```python
v0l_bin = open(dirin+v0lowfil, 'rb')
tmpv0l = v0l_bin.read(recl)
v0l = np.reshape(array('f',tmpv0l),(Ny,Nx))
```

```python
h0l_bin = open(dirin+h0lowfil, 'rb')
tmph0l = h0l_bin.read(recl)
h0l = np.reshape(array('f',tmph0l),(Ny,Nx))
```

```python
fig, ax = plt.subplots(figsize=(15,5))
plt.xlabel('Nx gridpoints', fontsize=16)
plt.ylabel('Ny gridpoints', fontsize=16)
plt.suptitle('Shallow Water Model Initial Wind (Barbs) & Height (Contours), d = 500km', fontsize=20)
x = np.arange(Nx)
y = np.arange(Ny)
h0lc = ax.contour(x,y,h0l)
wind0l = ax.quiver(x,y,u0l,v0l,color='black',headwidth=1,headlength=3)
```
