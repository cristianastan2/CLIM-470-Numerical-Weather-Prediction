```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
```

```python
data=pd.read_csv('NWPHW2DATA.csv')
data.columns=['Day', 'Tmin Forecast', 'Tmin Verified']
data=data.set_index('Day')
for col in ['Tmin Forecast', 'Tmin Verified']:
    data[col]=pd.to_numeric(data[col])
```

```python
plt.figure(figsize=(20,10))
plt.plot(data.index, data['Tmin Forecast'], color='r', linewidth=2.0)
plt.plot(data.index, data['Tmin Verified'], color='b', linewidth=2.0)
plt.xlabel('Day', fontsize=16)
plt.ylabel('Temperature (F)', fontsize=16)
plt.suptitle('Tmin Forecast vs. Verification Using McKenzie Method', fontsize=20)
red_patch = mpatches.Patch(color='red', label='Tmin Forecast')
blue_patch = mpatches.Patch(color='blue', label='Tmin Verified')
plt.legend(handles=[red_patch,blue_patch])
```
