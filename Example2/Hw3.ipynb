```python
import numpy as np
import matplotlib.pyplot as plt

# Define RH range
RH = np.linspace(0, 1.1, 500)

# Define the parameterization function
def cloud_coverage(RH, RH0):
    C = np.zeros_like(RH)
    mask1 = (RH > RH0) & (RH < 1)
    C[mask1] = ((RH[mask1] - RH0) / (1 - RH0))**2
    C[RH >= 1] = 1
    return C

# Parameters
RH0_lowhigh = 0.8
RH0_mid = 0.65

# Compute C for both cases
C_lowhigh = cloud_coverage(RH, RH0_lowhigh)
C_mid = cloud_coverage(RH, RH0_mid)

# Plot
plt.figure(figsize=(8,5))
plt.plot(RH, C_lowhigh, label=r'$RH_0 = 0.8$ (low/high clouds)', linewidth=2)
plt.plot(RH, C_mid, label=r'$RH_0 = 0.65$ (mid-level clouds)', linewidth=2, linestyle='--')

plt.xlabel('Relative Humidity (RH)')
plt.ylabel('Cloud Coverage (C)')
plt.title('Parameterized Cloud Coverage vs Relative Humidity')
plt.legend()
plt.grid(True)
plt.ylim(0, 1.05)
plt.xlim(0, 1.05)
plt.show()
```

```python
H = np.logspace(-5, 0, 500)
x = 1
actual = np.cos(1)

def first(x,H):
    approx = ((np.sin(x+H) - np.sin(x)) / H)
    error = np.absolute(approx - actual)
    return (error) 

def second(x,H):
    approx = ((np.sin(x+H)-np.sin(x-H)) / (2*H)) 
    error = np.absolute(approx - actual)
    return (error)

first_order = first(x,H)
second_order = second(x,H)
    
# Plot
plt.figure(figsize=(8,5))
plt.plot(H, first_order, label=r'$first_order', linewidth=2)
plt.plot(H, second_order, label=r'$second_order', linewidth=2, linestyle='--')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Delta x')
plt.ylabel('Error')
plt.title('Delta x vs Error')
plt.legend()
plt.grid(True)
plt.show()
```
