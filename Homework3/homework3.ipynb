```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
```

```python
data=pd.read_csv('q1.csv')
data.columns=['RH','C(RH0=0.8)','C(RH0=0.65)']
data=data.set_index('RH')
for col in ['C(RH0=0.8)','C(RH0=0.65)']:
    data[col]=pd.to_numeric(data[col])
```

```python
plt.figure(figsize=(20,10))
plt.plot(data.index, data['C(RH0=0.8)'], color='r', linewidth=2.0)
plt.plot(data.index, data['C(RH0=0.65)'], color='b', linewidth=2.0)
plt.xlabel('Relative Humidity', fontsize=16)
plt.ylabel('Cloud Coverage', fontsize=16)
plt.suptitle('Parameterized Cloud Coverage vs. Relative Humidity', fontsize=20)
red_patch = mpatches.Patch(color='red', label='C(RH0=0.8)')
blue_patch = mpatches.Patch(color='blue', label='C(RH0=0.65)')
plt.legend(handles=[red_patch,blue_patch])
```

```python
data=pd.read_csv('q2.csv')
data.columns=['delx','logdelx','Euler','AdamsBashforth','LogAbsErrEuler','LogAbsErrAdamsBashforth']
data=data.set_index('logdelx')
for col in ['LogAbsErrEuler','LogAbsErrAdamsBashforth']:
    data[col]=pd.to_numeric(data[col])
```

```python
plt.figure(figsize=(20,10))
plt.plot(data.index, data['LogAbsErrEuler'], color='r', linewidth=2.0)
plt.plot(data.index, data['LogAbsErrAdamsBashforth'], color='b', linewidth=2.0)
plt.xlabel('Log of Delta x', fontsize=16)
plt.ylabel('Log of Absolute Value of Error', fontsize=16)
plt.suptitle('First & Second Order Schemes to Approximate df/dx, f(x)=cos(x), x1= pi', fontsize=20)
red_patch = mpatches.Patch(color='red', label='Euler Method')
blue_patch = mpatches.Patch(color='blue', label='Adams-Bashforth Scheme')
plt.legend(handles=[red_patch,blue_patch])
```
